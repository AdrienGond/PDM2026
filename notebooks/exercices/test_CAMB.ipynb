```python
import numpy as np
import matplotlib.pyplot as plt
import camb
import os
```

```python
print('Using CAMB %s installed at %s' % (camb.__version__, os.path.dirname(camb.__file__)))
```

```python
# Set up a new set of parameters for CAMB
# The defaults give one massive neutrino and helium set using BBN consistency
pars = camb.set_params(
    H0=67.5, ombh2=0.022, # baryon density
    omch2=0.122,# cold dark matter density
    mnu=0.06, # sum of neutrino masses (in eV)
    omk=0, # Omega_K curvature parameter
    #tau=0.06,
    As=2e-9,
    ns=0.965,
    halofit_version='mead', # name of the specific Halofit model to use for non-linear modelling
    lmax=3000 # lmax for accuracy settings
)
```

```python
# calculate results for these parameters
results = camb.get_results(pars)
```

```python
# get dictionary of CAMB power spectra
powers = results.get_cmb_power_spectra(pars, CMB_unit='muK')
for name in powers:
    print(name)
```

```python
rs = results.sound_horizon(1080)  # sound horizon at redshift of last scattering
print('Sound horizon at last scattering: %.3f Mpc' % rs)
```

```python
# params.set_cosmology(H0=H0, ombh2=Omb*h**2, omch2=Omc*h**2, mnu=0.06, omk=0, tau=0.06)
params.set_cosmology(H0=H0, ombh2=0.0224, omch2=0.12084, mnu=0.06, omk=0, tau=0.06)
w0 = -0.7
wa = -1.0
params.set_dark_energy(w=w0, wa=wa, dark_energy_model='ppf')
```

```python
results = camb.get_results(params)
```
