### mean

```python
for z_index in range(1,7):
    labels_alt_ = [labels_alt[i] for i in range(7) if i != z_index]
    # print(labels_alt_)

    redshifts = []
    data = []
    data_types = []

    filenames = os.listdir(path)
    for i in range(len(labels_alt_)):
        for f in filenames:
            if labels_alt_[i] in f:
                if "mean" in f:
                    dat = np.loadtxt(path + "\\" + f, dtype=str)
                    if labels_alt[0] in f:
                        redshifts.append(float(dat[0]))
                        data.append(float(dat[1]))
                        data_types.append(dat[2])
                    else:
                        for j in range(len(dat)):
                            redshifts.append(float(dat[j][0]))
                            data.append(float(dat[j][1]))
                            data_types.append(dat[j][2])

    all_but_z_mean = [[redshifts[i], data[i], data_types[i]] for i in range(len(redshifts))]
    np.savetxt("all_but_z" + str(z_index+1) + "_mean.txt", all_but_z_mean, fmt="%s")
```
